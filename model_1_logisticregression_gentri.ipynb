```python
%matplotlib inline
import matplotlib
import pandas as pd
import os
import numpy as np

# to see in class example refer to 21/2/02
```

# Read the CSV

```python
df = pd.read_csv("gentrified_LA_data01.csv")
# Drop the null columns where all values are null
# df = df.dropna(axis='columns', how='all')
# Drop the null rows
# df = df.dropna()
df.head()
```

# Select control model features, place into dataframe

```python
# Set up x values/features df with just control model columns, this is cut #1

# show all columns
pd.set_option('display.max_columns', None)

# reduce set to cols 0-29
selected_features = df.iloc[:, 0:29]
selected_features.head()
```

```python
# get the shape of the file
selected_features.shape
```

```python
# remove non-feature columns
selected_features = selected_features.drop(['OBJECTID', 'Zip Code', 'Rank', 'Neighborhood', 'Index Score', 'Index Percentile', '% Change in Median Gross Rent'], axis=1)
selected_features.head()
```

```python
# get the shape of the file
selected_features.shape
```

### Assign x and y variables

```python
# assign X (data) and y (target)

# note we drop our outcome variable from the rest of the data to get our X variables
X = selected_features.drop("Index Quartile", axis=1)

# define our y by selecting the column within the dataframe
y = selected_features["Index Quartile"]

# get # of observations and variables (note, the 2nd parens doesn't have a variable number
# because the values is still a string)
print(X.shape, y.shape)
```

### Split the data into training and testing

```python
from sklearn.model_selection import train_test_split

# i modded this train_size
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, train_size=0.65, stratify=y)
```

```python
# confirm size of training set (default is 75% of the dataframe)
len(X_train)
```

```python
# confirm size of test set (the other 25%)
len(X_test)
```

### Create a logistic regression model

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
 LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
          intercept_scaling=1, max_iter=100, multi_class='ovr', n_jobs=1,
          penalty='l2', random_state=None, solver='liblinear', tol=0.0001,
          verbose=0, warm_start=False)
```

### Fit (train) or model using the training data

```python
classifier.fit(X_train, y_train)
```

```python
 LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
          intercept_scaling=1, max_iter=100, multi_class='ovr', n_jobs=1,
          penalty='l2', random_state=None, solver='liblinear', tol=0.0001,
          verbose=0, warm_start=False)
```

### Validate the model using the test data

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

### Make predictions

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
```
