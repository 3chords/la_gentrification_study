```python
%matplotlib inline
import matplotlib
import pandas as pd
import os

# to see in class example refer to 21/2/02
```

# Read the CRE CSV

```python
df = pd.read_csv("gentrified_LA_data02.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()

# show all columns
pd.set_option('display.max_columns', None)

df.head()
```

```python
# get the shape of the file
df.shape
```

# Select CRE features, place into dataframe

```python
# remove non-feature columns
selected_features = df.drop(['OBJECTID'], axis=1)
selected_features.head()
```

```python
# get the shape of the file
selected_features.shape
```

```python
# run correlation of CRE variables against index score
selected_features.corr()
```

### Keep CRE columns with strongest correlations

```python
# keep higher correlation columns
selected_features = selected_features[['Zip Code', 'Index Quartile', 'Total Retail Bldgs 2014', 'New MF Projects 2000-14', \
                                      '% Change in Retail Bldgs 2000-14', '% Studio', '% 2BR']]
selected_features.head()
```

```python
# get the shape of the file
selected_features.shape
```

### Bring in Census data

```python
# show all columns
pd.set_option('display.max_columns', None)

df_census = pd.read_csv("gentrified_LA_data01.csv")
df_census.head()
```

```python
# get the shape of the file
df_census.shape
```

```python
# Set up x values/features df with just control model columns, this is cut #1

# rename dataframe
df_census = df_census.iloc[:, 0:29]
df_census.head()
```

```python
# get the shape of the file
df_census.shape
```

```python
# remove non-feature columns
df_census = df_census.drop(['OBJECTID', 'Neighborhood', 'Rank', 'Index Score', 'Index Quartile'], axis=1)
df_census.head()
```

```python
# get the shape of the file
df_census.shape
```

### Merge CRE with Census file

```python
selected_features_CRE = pd.merge(df_census, selected_features, on='Zip Code')
selected_features_CRE.head()
```

```python
# get the shape of the file
selected_features_CRE.shape
```

```python
# write file to csv (note this includes the x variable and the zip code, use this for mapping)

selected_features_CRE.to_csv("census_cre_features_w_zip.csv", index=False, header=True)
```

```python
# remove non-feature columns
selected_features_CRE = selected_features_CRE.drop(['Zip Code', 'Index Percentile'], axis=1)
selected_features_CRE.head()
```

```python
# get the shape of the file
selected_features_CRE.shape
```

### Assign x and y variables

```python
# one-hot encode just the index quartile column, we want to change the coding to just reveal the top 25% = 1, rest=0
selected_features_CRE = pd.get_dummies(selected_features_CRE, prefix=['Index Quartile'], columns=['Index Quartile'])
selected_features_CRE.head()
```

```python
# assign X (data) and y (target)

# note we drop our new outcome variable from the rest of the data to get our X variables
X = selected_features_CRE.drop("Index Quartile_1", axis=1)

# define our y by selecting the column within the dataframe
y = selected_features_CRE["Index Quartile_1"]

# get # of observations and variables (note, the 2nd parens doesn't have a variable number
# because the values is still a string)
print(X.shape, y.shape)
```

### Split the data into training and testing

```python
from sklearn.model_selection import train_test_split

# i modded this train_size
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, train_size=0.65, stratify=y)
```

```python
# confirm size of training set (default is 75% of the dataframe)
len(X_train)
```

```python
# confirm size of test set (the other 25%)
len(X_test)
```

### Scale the variables

```python
from sklearn.preprocessing import StandardScaler

X_scaler = StandardScaler().fit(X_train)
```

```python
# scale both the training and the testing data
X_train = X_scaler.transform(X_train)
X_test = X_scaler.transform(X_test)
```

### Create a logistic regression model

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
 LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
          intercept_scaling=1, max_iter=100, multi_class='ovr', n_jobs=1,
          penalty='l2', random_state=None, solver='liblinear', tol=0.0001,
          verbose=0, warm_start=False)
```

### Fit (train) or model using the training data

```python
classifier.fit(X_train, y_train)
```

```python
 LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
          intercept_scaling=1, max_iter=100, multi_class='ovr', n_jobs=1,
          penalty='l2', random_state=None, solver='liblinear', tol=0.0001,
          verbose=0, warm_start=False)
```

### Validate the model using the test data

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

### Make predictions

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
```

```python
# get coefficients, now because they are scaled we can truly assess weight of the their influence on the model
print(classifier.coef_)
```
